```
lyst=[Tag for Tags in Finaldf['Tag'].values for Tag in Tags]
```

```
dlyst=nltk.FreqDist(lyst)
```

```
dlyst=dict(dlyst)
```

```
v=sorted(dlyst.items(),reverse=True,key=lambda x: x[1])
```

Here I am taking Top 10 tags only 


```
top_10=[c[0] for c in v][0:10]
```

```
def most_occured(tags):
    filterr = []
    for i in range(0, len(tags)):
        if tags[i] in top_10:
            filterr.append(tags[i])
    return filterr
```

```
def preprocessing(b):

  b=str(b)
  b=b.lower()
  b=re.sub(r"what's",'what is',b )
  b=re.sub(r"i've",'i have',b)
  b=re.sub(r"can't",'can not',b)
  b=re.sub(r"\'\n","  ",b)
  b=re.sub(r"\'s",' is',b)
  b=re.sub(r"\'ve"," have",b)
  b=re.sub(r"\'r",' are',b)
  b=re.sub(r"\'m",' am',b)
  b=re.sub(r"\'d",' would',b)
  b=re.sub(r"\'cause"," because",b)
  b=re.sub(r"\'\n","   ",b)
  b=b.strip(" ")
  return b
```

```
Finaldf["Body"]=Finaldf["Body"].apply(lambda x: preprocessing(x))
```

```
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import ToktokTokenizer
from nltk.corpus import stopwords
from string import punctuation
from nltk.tokenize import word_tokenize
lemma = WordNetLemmatizer()
my_stop=set(stopwords.words('english')+list(punctuation))
```

```
Tok=ToktokTokenizer()
```

```
top_10
```

```
punct = '!"#$%&\'()*+,./:;<=>?@[\\]^_`{|}~.'

```

```
def split_into_lemmas(message):
    message=message.lower()
    words = Tok.tokenize(message)
    regex = re.compile('[%s]' % re.escape(punct))
    words_sans_stop=[]
    for word in words :
      if word in top_10:
        words_sans_stop.append(word)
      elif word not in my_stop:
        word = lemma.lemmatize(word)
        words_sans_stop.append(regex.sub(' ', word))
    return '  '.join(map(str, words_sans_stop))
```

```
Finaldf['Body']= Finaldf['Body'].apply(lambda x: split_into_lemmas(x))
```

```

tok = Tokenizer(char_level=False,split=' ')
tok.fit_on_texts(X_train['Combine'])
sequences_train = tok.texts_to_sequences(X_train['Combine'])
sequences_test=tok.texts_to_sequences(X_test['Combine'])
```

```
vocab_len=len(tok.index_word)
```

```
from keras.preprocessing import sequence
```

```
sequences_matrix_train = sequence.pad_sequences(sequences_train,maxlen=235)
sequences_matrix_test = sequence.pad_sequences(sequences_test,maxlen=235)
```

```
sequences_matrix_train.shape
```

```
from keras.layers import LSTM, Activation, Dense, Dropout, Input, Embedding
from keras.models import Model
```

```
max_len=235
```

```
def RNN():
    inputs = Input(name='inputs',shape=[max_len])
    layer = Embedding(vocab_len+1,300,input_length=max_len,mask_zero=True)(inputs)
    layer = LSTM(64)(layer)
    layer = Dense(128,name='FC1')(layer)
    layer = Activation('relu')(layer)
    layer = Dropout(0.5)(layer)
    layer = Dense(10,name='out_layer')(layer)
    layer = Activation('sigmoid')(layer)
    model = Model(inputs=inputs,outputs=layer)
    return model
```

```
model=RNN()
model.summary()
model.compile(loss='binary_crossentropy',optimizer='adam',
              metrics=['accuracy'])
```

```
model.fit(sequences_matrix_train,y_train,batch_size=500,epochs=10,validation_data=(sequences_matrix_test,y_test))
```
